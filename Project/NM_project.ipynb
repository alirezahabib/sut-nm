**Theoretical:**
<br>Let $a_i$ ($1 \le i \le n$) be a positive number. Prove that

a): $e(\sum_{i}^{n}{a_i}) \le \sum_{i}^{n}{e(a_i)} $

b): $\delta(\sum_{i}^{n}{a_i}) \le max\{\delta(a_1),...,\delta(a_n)\}$


Write your answers here:

---
### a

از استقرا استفاده می‌کنیم
          
$e(\sum_{i}^{n}{a_i}) = e(\sum_{i}^{n - 1}{a_i} + a_n) = e(s + a_n)$

$s := \sum_{i}^{n - 1}{a_i}$

$|s - \bar{s}| \leq e(s), \quad |a_n - \bar{a_n}| \leq e(a_n)$

$\bar{s} - e(s) \leq s \leq \bar{s} + e(s), \quad \bar{a_n} - e(a_n) \leq a_n \leq \bar{a_n} + e(a_n)$

$\bar{s} + \bar{a_n} - e(s) - e(a_n) \leq s + a_n \leq \bar{s} + \bar{a_n} + e(s) + e(a_n)$

$|s + a_n - \bar{s} - \bar{a_n}| \leq e(s) + e(a_n)$

$\implies e(s + a_n) \leq e(s) + e(a_n)$

$\implies e(\sum_{i}^{n}{a_i}) \leq e(s) + e(a_n)$

اما از حکم استقرا برای
$n - 1$
می‌دانیم:

$e(s) = e(\sum_{i}^{n - 1}{a_i}) \leq \sum_{i}^{n-1}{e(a_i)}$

$\implies e(\sum_{i}^{n}{a_i}) \leq [\sum_{i}^{n-1}{e(a_i)}] + e(a_n) = \sum_{i}^{n}{e(a_i)}$

برای پایه‌ی استقرا
$n = 1$
خوب است چون حکم بدیهی است و گام استقرا هم برای
$n = 1$
به 
$n = 2$
مشکلی ندارد.

$e(\sum_{i}^{1}{a_i}) = e(a_1) \leq \sum_{i}^{1}{e(a_i)} = e(a_1)$

$\blacksquare$

### b
فرض کنید
$\sum_{i}^{n}{a_i}$
عددی نزدیک به صفر باشد ولی خود
$a_i$ها
مقادیر دور از صفر
مثبت و منفی.
در این حالت خطای نسبی جمع آن‌ها
یعنی
$\delta(\sum_{i}^{n}{a_i})$
به مقدار دلخواه بزرگ می‌شود اما
هر کدام از خطاهای نسبی متغییرها محدود می‌ماند.
پس این یک مثال نقض است و 
$\delta(\sum_{i}^{n}{a_i})$
در این مثال نمی‌تواند کمتر از
$max\{\delta(a_1),...,\delta(a_n)\}$
باشد.

حکم برای مقادیر حقیقی
$a_i$
لزوما درست نیست ولی برای مقادیر مثبت یعنی
$a_i > 0$
درست است:

$\delta(\sum_{i}^{n}{a_i}) = \frac{e(\sum_{i}^{n}{a_i})}{|\sum_{i}^{n}{a_i}|} = \frac{e(\sum_{i}^{n}{a_i})}{\sum_{i}^{n}{a_i}} \leq \frac{\sum_{i}^{n}{e(a_i)}}{\sum_{i}^{n}{a_i}}$

که در بالا از نتیجه‌ی بخش قبل استفاده کردیم.

$e(a_i) = a_i \delta(a_i)$

$\implies \frac{\sum_{i}^{n}{e(a_i)}}{\sum_{i}^{n}{a_i}} = \frac{\sum_{i}^{n}{a_i \delta(a_i)}}{\sum_{i}^{n}{a_i}} \leq \frac{\sum_{i}^{n}{a_i \delta(a_m)}}{\sum_{i}^{n}{a_i}} = \frac{[\sum_{i}^{n}{a_i}] \delta(a_m)}{\sum_{i}^{n}{a_i}} = \delta(a_m)$

$\delta(a_m) := max\{\delta(a_1),...,\delta(a_n)\}$

$\implies \delta(\sum_{i}^{n}{a_i}) \leq max\{\delta(a_1),...,\delta(a_n)\}$

$\blacksquare$

  a = Num(1.23456)

````python
import math

class Num():
    '''
      complete the class so that there's truncation when
      creating a Number and correct error propagation
      when using operations on Numbers
      Hint: you must use operator overloading
    '''
    def __init__(self, val):
        self.value = round(val, 2)
        self.abs_err = abs(val - self.value)
        self.rel_err = self.abs_err / abs(val)

    def __add__(self, other):
        exact = self.value + other.value
        ans = Num(exact)
        ans.abs_err = self.abs_err + other.abs_err
        ans.rel_err = ans.abs_err / abs(exact)
        return ans

    def __sub__(self, other):
        exact = self.value - other.value
        ans = Num(exact)
        ans.abs_err = self.abs_err + other.abs_err
        ans.rel_err = ans.abs_err / abs(exact)
        return ans

    def __mul__(self, other):
        exact = self.value * other.value
        ans = Num(exact)
        # Unlike sum and sub, mul and div can have errors in rounding in addition to propagation
        # (1.11 ± 0.00) * (1.11 ± 0.00) = (1.23 ± 0.002)
        # note the "+=" here
        ans.abs_err += abs(self.value) * self.abs_err + abs(other.value) * other.abs_err
        ans.rel_err = ans.abs_err / abs(exact)
        return ans

    def __truediv__(self, other):
        exact = self.value / other.value
        ans = Num(exact)
        # Same explanation for "+=" as the mul function
        ans.abs_err += self.abs_err / abs(other.value) + abs(self.value) * other.abs_err / other.value ** 2
        ans.rel_err = ans.abs_err / abs(exact)
        return ans


a = Num(1.23456)
b = Num(5.67891)
c = a + b
print(a.value)
print(a.abs_err)
print(a.rel_err)
print(c.value)
````

     ```
x * x + y * y - 4
x * y -  1
     ```
     <br>
     <br>

**Hint: see last slides of lecture two for more information.**

````python
import sympy as sp
from sympy.abc import x, y

F = sp.parse_expr(input("F(x,y): "))
G = sp.parse_expr(input("G(x,y): "))
#F = x * x + y * y - 4
#G = x * y - 1

Fx = sp.diff(F, x)
Fy = sp.diff(F, y)
Gx = sp.diff(G, x)
Gy = sp.diff(G, y)

D = Fx * Gy - Fy * Gx
D1 = -F * Gy + G * Fy
D2 = -G * Fx + F * Gx

x_val = float(input("x_0: "))
y_val = float(input("y_0: "))
#x_val = 2
#y_val = 0
F_val = 1
G_val = 1
error = 0.0001

while abs(F_val) > error or abs(G_val) > error:
    F_val = F.subs([(x, x_val), (y, y_val)])
    G_val = G.subs([(x, x_val), (y, y_val)])
    D_val = D.subs([(x, x_val), (y, y_val)])
    if D_val == 0:
        raise "No solution"
    D1_val = D1.subs([(x, x_val), (y, y_val)])
    D2_val = D2.subs([(x, x_val), (y, y_val)])
    x_val += D1_val / D_val
    y_val += D2_val / D_val

print(f'x = {x_val.evalf(6)}, y = {y_val.evalf(6)}')
print(f'F(x,y) = {F_val.evalf(6)}, G(x,y) = {G_val.evalf(6)}')
````

<h2>Simple Linear Regression (40 points) </h2>
<p>
In statistics, simple linear regression is a linear regression model with a single explanatory variable.[1][2][3][4][5] That is, it concerns two-dimensional sample points with one independent variable and one dependent variable (conventionally, the x and y coordinates in a Cartesian coordinate system) and finds a linear function (a non-vertical straight line) that, as accurately as possible, predicts the dependent variable values as a function of the independent variable.
The formula for this method is:
Y = β_0 + β_1 * X + ϵ
Where
Y is the predicted value of the dependent variable (y) for any given value of the independent variable (x).
β_0 is the intercept, the predicted value of y when the x is 0.
β_1 is the regression coefficient – how much we expect y to change as x increases.
X  is the independent variable ( the variable we expect is influencing y).
ϵ is the error of the estimate, or how much variation there is in our estimate of the regression coefficient.
</br>
Linear regression finds the line of best fit through your data by searching for the regression coefficient (β_1) that minimizes the total error (ϵ) of the model according to least square error.
</p>

````python
def find_coefficients(feature_vector, response_vector):
    cov = np.cov(feature_vector, response_vector)
    b = cov[0, 1] / cov[0, 0] # = cov(x, y) / cov(x, x)
    a = np.mean(response_vector) - b * np.mean(feature_vector)
    return a, b


def plot_result(feature_vector, response_vector, b_0, b_1):
    plt.scatter(feature_vector, response_vector)
    plt.plot(feature_vector, b_0 + b_1 * feature_vector, 'r')
    plt.title('Linear Regression')
    plt.xlabel('feature')
    plt.ylabel('response')
    plt.show()
````

<h2> Discrete Fourier Transform (DFT) (25 points)</h2>
<p>
In mathematics, the discrete Fourier transform (DFT) converts a finite sequence of equally-spaced samples of a function into a same-length sequence of equally-spaced samples of the discrete-time Fourier transform (DTFT), which is a complex-valued function of frequency. The interval at which the DTFT is sampled is the reciprocal of the duration of the input sequence.

The formula for this transformation is:

$X_{k} = \sum_{n = 0}^{N - 1} x_{n} . e^{i 2\pi k n /N} = \sum_{n = 0}^{N - 1}x_{n}\left [ cos(2\pi k n/N) - i.sin(2\pi k n/N)) \right ]$

where

N = number of samples

n = current sample

k = current frequency, where k∈[0,N−1]

xn = the sine value at sample n

Xk = The DFT which includes information on both amplitude and phase
</p>

<p>
Here, we have caculated sum of three trigonometric functions and plotted them as one function. Implement a function to calculate DFT for a function which is the summation of three trigonometric functions.

````python
plt.style.use('seaborn-poster')
%matplotlib inline
def plot_function(sampling_rate):
# sampling rate
    # sampling interval
    interval = 1.0/sampling_rate
    t = np.arange(0,1,interval)

    freq = 2
    x = 5*np.sin(2*np.pi*freq*t)

    freq = 0.5
    x += -1 * np.cos(2*np.pi*freq*t)

    freq = 15  
    x += 2.5* np.sin(2*np.pi*freq*t)

    plt.figure(figsize = (8, 6))
    plt.plot(t, x, 'g')
    plt.ylabel('Amplitude')

    plt.show()
    return x
````

````python
def DFT(x):
    # 25 points
    """
    Function to calculate the 
    DFT of a 1D real-valued signal x
    """
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    return np.dot(x, np.exp(-2j * np.pi * k * n / N))
````

````python
def plot_DFT_or_FFT(X, sampling_rate, plot_type):
    # calculate the frequency
    N = len(X)
    n = np.arange(N)
    T = N/sampling_rate
    freq = n/T 

    plt.figure(figsize = (8, 6))
    plt.stem(freq, abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
    plt.xlabel('Freq (Hz)')
    label = plot_type + 'Amplitude |X(freq)|'
    plt.show()
````

````python
sampling_rates = [4, 16, 64, 128, 256, 512]
for sampling_rate in sampling_rates:
    x = plot_function(sampling_rate)
    X = DFT(x)
    plot_DFT_or_FFT(X, sampling_rate, 'DFT')
````

````python
def FFT(x):
    # 20 points
    """
    A recursive or  non-recursive implementation of 
    the 1D Cooley-Tukey FFT, the 
    input should have a length of 
    power of 2. 
    """
    N = len(x)
    
    if N == 1:
        return x
    
    X_e = FFT(x[::2])
    X_o = FFT(x[1::2])
    
    e = np.exp(-2j * np.pi * np.arange(N // 2) / N)
    return np.concatenate([X_e + e * X_o, X_e - e * X_o])
````

````python
# 5 points

import time

n = len(sampling_rates)


def function_silent(sampling_rate):
    interval = 1.0/sampling_rate
    t = np.arange(0,1,interval)

    freq = 2
    x = 5*np.sin(2*np.pi*freq*t)

    freq = 0.5
    x += -1 * np.cos(2*np.pi*freq*t)

    freq = 15  
    x += 2.5* np.sin(2*np.pi*freq*t)

    return x


def test():
    DFT_times = np.zeros(n)
    FFT_times = np.zeros(n)

    for i, sampling_rate in enumerate(sampling_rates):
        x = function_silent(sampling_rate)

        t0 = time.time()
        DFT(x)
        t1 = time.time()
        FFT(x)
        t2 = time.time()

        DFT_times[i] = (t1 - t0) * 1000
        FFT_times[i] = (t2 - t1) * 1000
        
    return DFT_times, FFT_times


DFT_average = np.zeros(n)
FFT_average = np.zeros(n)

k = 10
for i in range(k):
    t_d, t_f = test()
    DFT_average += t_d
    FFT_average += t_f
    
DFT_average /= n
FFT_average /= n

for i in range(n):
    print(f'function {i}, DFT average time: {DFT_average[i]:7.5f}ms, FFT average time: {FFT_average[i]:7.5f}ms')
````
